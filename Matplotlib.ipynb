# scatter plot

```python
'''
now to create a scatter plot we need two arrays 
'''
a = np.random.randint(1,20,(1,10))
b = np.random.randint(1,20,(1,10))

# scatter plot
plt.scatter(a,b)
```

```python
#adding more value to our graph
plt.scatter(a,b,c='red')
plt.xlabel('array a')
plt.ylabel('array b')
plt.title('Simple 2D graph')
```

# plot

```python
x = np.arange(0,20)
y = x**3

plt.plot(x,y)
```

```python
# adding more features
plt.plot(x,y,'*',c='red',linestyle='dashed',linewidth=0.5)
plt.xlabel('Array x',c='blue')
plt.ylabel('Array y',c='green')
plt.title('plot')
```

```python
# creating subplots in a single canvas

plt.subplot(3,3,1) # in this line I am telling that I want a 3x3 canvas and at first index apply the bellow properties 
plt.plot(x,y,'r--')
plt.subplot(3,3,3)
plt.plot(x,y,'b--')
plt.subplot(3,3,7)
plt.plot(x,y,'g--')
plt.subplot(3,3,9)
plt.plot(x,y,'y--')
```

```python
# sin plot
x = np.arange(0, 9*np.pi,0.1)
y = np.sin(x)

plt.plot(x,y)
```

```python
# cas plot
x = np.arange(0, 9*np.pi,0.1)
y = np.cos(x)
plt.plot(x,y)
```

```python
# tan plot
x = np.arange(0, 9*np.pi,0.1)
y = np.tan(x)

plt.plot(x,y)
```

# bar plot

```python
x = np.arange(0,10)
y = np.random.random(10)  

plt.bar(x,y,color='green') 
#width attribute tells the width of each bar 
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Bar plot')
```

# box plot

```python
x = np.random.random(10)
plt.boxplot(x,patch_artist=True,) 
# patch_artist fills the color in box
```

# histogram

```python
x = np.random.random(3)
y = np.random.random(5)

plt.hist(x,bins=15,color='red',edgecolor='black')
plt.hist(y,bins=15,color='blue')
#edge color creates outline outside the bar
```

# pie chart

```python
y = ['Hardwork','consitency','luck','prayers']
x = [45,45,5,5] 
exp = [0,0,0,0.5] # for out from circle
plt.pie(x,labels=y,explode=exp)
```
